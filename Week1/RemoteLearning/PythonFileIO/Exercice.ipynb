```python
# Reading the file safely
with open("nameslist.txt", encoding='utf-8') as f:
    lines = f.readlines()                   # Read all lines into a list
    print("All lines:\n", lines)

    # 1. Read only the 5th line (index 4)
    if len(lines) >= 5:
        print("\nFifth line:", lines[4])
    else:
        print("\nThe file has fewer than 5 lines.")

    # 2. Read only the first 5 characters of the file
    f.seek(0)
    first_five = f.read(5)
    print("\nFirst 5 characters of the file:", first_five)

# 3. Combine all lines into a single string
file_string = "".join(lines)
print("\nCombined file string:\n", file_string)

# 4. Split the combined string into a list of words
words = file_string.split()
print("\nSplit words:\n", words)

#5. Find out how many occurences of the names "Darth", "Luke" and "Lea" are in the file
# Names to count
target_names = ["Darth", "Luke", "Lea"]
name_counts = {name: 0 for name in target_names}

# Read and process the file
with open("nameslist.txt", encoding='utf-8') as f:
   text = f.read()

# Split into words (basic split on whitespace)
words = text.split()

# Count occurrences
for word in words:
   for name in target_names:
      if word == name:
            name_counts[name] += 1

# Show the result
print("Name occurrences:")
for name, count in name_counts.items():
   print(f"{name}: {count}")

#6. Append your first name at the end of the file
with open("nameslist.txt", "a", encoding="utf-8") as f:
    f.write("\nJaled")
# Confirm it was added
with open("nameslist.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Print the last line of the file to confirm
print("\nLast lines of the file:")
for line in lines[-1:]:
    print(line.strip())

# 7. Append "Skywalker" next to each occurrence of "Luke"
with open("nameslist.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Modify each line: append ' Skywalker' after 'Luke' if it's found
updated_lines = []
for line in lines:
    if "Luke" in line:
        # Replace "Luke" with "Luke Skywalker", just once per line
        updated_line = line.replace("Luke", "Luke Skywalker")
        updated_lines.append(updated_line)
    else:
        updated_lines.append(line)

# Write modified lines back to the file
with open("nameslist.txt", "w", encoding="utf-8") as f:
    f.writelines(updated_lines)

with open("nameslist.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    print(lines)

```
