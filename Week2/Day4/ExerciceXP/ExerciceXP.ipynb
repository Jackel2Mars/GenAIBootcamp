Exercise 2: Creating A Line Plot For Temperature Variation

Objective: Create a simple line plot using Matplotlib that represents temperature variations over a week.
Tasks:
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]).

Exercise 4: Visualizing The Distribution Of CGPA

Objective: Create a histogram to visualize the distribution of students’ CGPA.
Dataset Overview: Assume the CGPA data is categorized into ranges and loaded in a DataFrame named df.
Tasks:
Import necessary libraries.
Use Seaborn’s histplot to create a histogram of the CGPA categories.
Customize the histogram with a specific color and add a title.
Display the plot.

```python
#Import necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np

# Expand data manually
raw_cgpa = []
ranges = ['6.0-6.9', '7.0-7.9', '8.0-8.9', '9.0-9.9']
counts = [12, 30, 45, 13]

for r, count in zip(ranges, counts):
    low, high = map(float, r.split('-'))
    raw_cgpa.extend(np.random.uniform(low, high, size=count))

# Plot histogram
sns.histplot(raw_cgpa, bins=4, kde=False, color='skyblue')

plt.title('Simulated CGPA Histogram')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.show()
```

Exercise 5: Comparing Anxiety Levels Across Different Genders

Objective: Use a bar plot to compare the proportion of students experiencing anxiety across different genders.
Dataset Overview: The dataset includes columns: ‘Do you have Anxiety?’ and ‘Choose your gender’.
Tasks:
Import necessary libraries.
Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.
Customize the plot with an appropriate color palette and add a title.
Display the plot.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate sample data
n = 200  # number of students
genders = np.random.choice(['Male', 'Female', 'Other'], size=n, p=[0.45, 0.45, 0.10])

# Simulate anxiety responses based on gender
anxiety_responses = []
for gender in genders:
    if gender == 'Female':
        anxiety_responses.append(np.random.choice(['Yes', 'No'], p=[0.7, 0.3]))
    elif gender == 'Male':
        anxiety_responses.append(np.random.choice(['Yes', 'No'], p=[0.5, 0.5]))
    else:
        anxiety_responses.append(np.random.choice(['Yes', 'No'], p=[0.6, 0.4]))

# Create DataFrame
df = pd.DataFrame({
    'Choose your gender': genders,
    'Do you have Anxiety?': anxiety_responses
})

# 1. Count the occurrences of each gender/anxiety pair
count_data = df.groupby(['Choose your gender', 'Do you have Anxiety?']).size().reset_index(name='Count')

# 2. Create the grouped bar plot
plt.figure(figsize=(8, 6))
sns.set(style='whitegrid')

sns.barplot(
    x='Choose your gender', 
    y='Count', 
    hue='Do you have Anxiety?',  # Different bars for anxiety 'Yes' and 'No'
    data=count_data, 
    palette='Set2'
)

# 3. Customize the plot
plt.title('Number of Students with and without Anxiety by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

Exercise 6: Exploring The Relationship Between Age And Panic Attacks

Objective: Create a scatter plot to explore the relationship between students’ age and the occurrence of panic attacks.
Dataset Overview: The dataset includes columns: ‘Age’ and ‘Do you have Panic Attacks?’.
Tasks:
Import necessary libraries.
Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
Use Seaborn’s scatterplot to create a scatter plot with ‘Age’ on the x-axis and numeric panic attack responses on the y-axis.
Customize the plot to improve readability by adding labels, a title, and adjusting point styles.
Display the plot.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# 1. Generate sample data
n = 200  # number of students
ages = np.random.randint(18, 30, size=n)  # random ages between 18 and 30
panic_responses = np.random.choice(['Yes', 'No'], size=n, p=[0.3, 0.7])  # 30% Yes, 70% No

# Create DataFrame
df = pd.DataFrame({
    'Age': ages,
    'Do you have Panic Attacks?': panic_responses
})

# 2. Convert panic attack responses to numeric values
df['Panic Attack Numeric'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})

# 3. Create a scatter plot
plt.figure(figsize=(8, 6))
sns.set(style='whitegrid')

sns.scatterplot(
    x='Age', 
    y='Panic Attack Numeric', 
    data=df, 
    hue='Panic Attack Numeric',  # Color points by panic attack occurrence
    palette='coolwarm',          # Color palette for better visual distinction
    style='Panic Attack Numeric', # Different markers for Yes (1) and No (0)
    markers={0: 'o', 1: 'X'},    # 'o' for No, 'X' for Yes
    s=100,                       # Size of the points
    alpha=0.7                    # Transparency of the points
)

# 4. Customize the plot
plt.title('Relationship Between Age and Panic Attacks', fontsize=16)
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Panic Attack Occurrence', fontsize=12)
plt.xticks(np.arange(18, 31, step=1))  # Set x-ticks from 18 to 30
plt.yticks([0, 1], labels=['No', 'Yes'])  # Label y-ticks as 'No' and 'Yes'
plt.tight_layout()

# 5. Display the plot
plt.show()
```
