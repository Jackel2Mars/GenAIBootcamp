1. Line Plot Exercise:

Create a line plot for the function y = x^2 for x ranging from 0 to 10. Add appropriate labels and a title.

```python
import matplotlib.pyplot as plt

x = range(10)
y = [i**2 for i in x]
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

2. Bar Chart Exercise:

Visualize the sales of five products (e.g., Product A, Product B, etc.) in a bar chart. Customize the chart with different colors and patterns.

```python
products = ['ProductA', 'ProductB', 'ProductC', 'ProductD', 'ProductE']
sales = [20, 35, 30, 15, 5]

plt.bar(products, sales, color=['red', 'green', 'blue', 'purple', 'orange'], edgecolor='black', hatch='/')
plt.title('Products Sales Data')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

3. Histogram Exercise:

Generate a dataset of 500 random numbers following a normal distribution. Plot the histogram and experiment with different bin sizes and colors.

```python
import numpy as np

data = np.random.normal(0, 1, 500)
plt.hist(data, bins=10, edgecolor='green', color='red')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

4. Pie Chart Exercise:

Create a pie chart showing the percentage distribution of time spent on different daily activities (e.g., work, sleep, leisure, exercise, etc.). Use custom colors and add a shadow effect.

```python
daily_activities = ['Work', 'Sleep', 'Leisure', 'Exercise', 'Others']
time_spent = [10, 6, 4, 3, 1]

plt.pie(time_spent, labels=daily_activities, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange', 'purple'], shadow=True)
plt.title('Percentage distribution of time spent on different daily activities')
plt.show()
```
