1. Create a histogram of customer spending from the tips dataset. Customize the bins and color.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
sns.histplot(data['tip'], bins=10, color='purple')
plt.title('Histogram of Tips')
plt.show()
```

2. Use pairplot on the iris dataset with different markers for species.

```python
iris_data = sns.load_dataset('iris')
sns.pairplot(iris_data, hue='species', palette='Dark2', markers=["o", "s", "D"])
plt.show()
```

Visualize correlations in the iris dataset using a heatmap.

```python
iris_numerical = iris_data.select_dtypes(include=['number']).columns
correlation_matrix = iris_data[iris_numerical].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Build a FacetGrid using the titanic dataset to compare survival rates by age across class and gender.

```python
titanic_data = sns.load_dataset('titanic')
titanic_data.info()
g = sns.FacetGrid(titanic_data, col='class', row='sex', margin_titles=True)
g.map(sns.scatterplot, 'age', 'survived', color='red')
plt.show()
```
