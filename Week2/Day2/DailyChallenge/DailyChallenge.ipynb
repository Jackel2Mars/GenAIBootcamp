```python
#For this challenge, please use google colab.
#Load and Visualize Images using the Flower Color Images dataset

from google.colab import files
files.upload()

!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

!kaggle datasets download -d olgabelitskaya/flower-color-images

!unzip flower-color-images.zip

#Install the libraries :
!pip install tensorflow
! pip install keras
! pip install pillow

# Load an image using PIL
from PIL import ImageEnhance, Image, ImageOps
import numpy as np
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
import matplotlib.pyplot as plt
#plt.imshow(original_image)

#Rotate an image by 90 degrees
from scipy.ndimage import rotate
def rotate_image_90_degrees(image):
    # Convert PIL image to NumPy array
    image_array = np.array(image)
    # Rotate by 90 degrees counter-clockwise (reshape=False keeps original size)
    rotated_array = rotate(image_array, angle=90, reshape=False)
    # Convert back to PIL image
    return Image.fromarray(rotated_array.astype(np.uint8))

rotated_90_image = rotate_image_90_degrees(original_image)

#Flip an image horizontally and then vertically using mirror and flip functions from ImageOps
# Flip horizontally (left-right)
horizontally_flipped = ImageOps.mirror(original_image)
# Flip vertically (top-bottom)
vertically_flipped = ImageOps.flip(original_image)

#Zoom in on an image (scale by 1.2x) using .resize(...)
def zoom_in(image, scale):
    width, height = image.size
    # Calculate crop size
    new_width = int(width / scale)
    new_height = int(height / scale)
    # Calculate coordinates for center crop
    left = (width - new_width) // 2
    top = (height - new_height) // 2
    right = left + new_width
    bottom = top + new_height
    # Crop the center of the image
    cropped_image = image.crop((left, top, right, bottom))
    # Resize back to original size (zoom effect)
    zoomed_image = cropped_image.resize((width, height), Image.LANCZOS)
    return zoomed_image
# Usage
zoomed_image = zoom_in(original_image, 1.2)

#Display all images
plt.figure(figsize=(12, 4))

plt.subplot(1, 5, 1)
plt.imshow(original_image)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(rotated_90_image)
plt.title("Rotated 90 degrees counter-clockwise")
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(horizontally_flipped)
plt.title("Flipped Horizontally")
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(vertically_flipped)
plt.title("Flipped Vertically")
plt.axis('off')

plt.subplot(1, 5, 5)
plt.imshow(zoomed_image)
plt.title("Zoomed in 1.2x")
plt.axis('off')


```
