```python
import numpy as np
# Movie ratings matrix (5 viewers, 3 movies)
movie_ratings = np.array([
    [5, 1, 4],
    [4, 4, 2],
    [4, 3, 5],
    [1, 1, 5],
    [3, 2, 1]
])

# Tasks: Calculate average ratings for each movie and analyze viewer preferences
average_ratings = np.mean(movie_ratings, 0)
print(average_ratings)
viewer_preferences = np.argmax(movie_ratings, 1)
print(viewer_preferences)
```

Scenario:

Analyze temperature data across four cities over a week.

Data Set:

The temperature data for 4 cities (City 1 to City 4) over 7 days (Day 1 to Day 7) is represented in the following matrix:


City 1	City 2	City 3	City 4	
27	24	23	22	# Day 1
20	18	23	29	# Day 2
15	20	27	23	# Day 3
18	17	25	16	# Day 4
26	19	16	20	# Day 5
18	22	21	24	# Day 6
22	21	22	28	# Day 7

Exercise Tasks:

1. Average Temperature Calculation: Compute the average temperature for each city over the week.

2. Maximum and Minimum Temperature Analysis: Find the highest and lowest temperatures recorded in each city.

3. Day Comparison: Compare the temperature across the cities on the first and last day of the week.


import numpy as np

# Temperature data matrix (4 cities, 7 days)
temperature_data = np.array([
    [27, 20, 15, 18, 26, 18, 22],
    [24, 18, 20, 17, 19, 22, 21],
    [23, 23, 27, 25, 16, 21, 22],
    [22, 29, 23, 16, 20, 24, 28]
])

# Tasks: Calculate average, max, min temperatures and make day comparisons

Expected Outputs:

Average Temperature: The average temperature for each city.
Max/Min Temperature: The highest and lowest temperatures in each city.
Day Comparison: Temperature differences between Day 1 and Day 7 in each city.

```python
# Temperature data matrix (4 cities, 7 days)
temperature_data = np.array([
    [27, 20, 15, 18, 26, 18, 22],
    [24, 18, 20, 17, 19, 22, 21],
    [23, 23, 27, 25, 16, 21, 22],
    [22, 29, 23, 16, 20, 24, 28]
])

#Average Temperature: The average temperature for each city.
average_temperatures = np.mean(temperature_data, 1)
print(average_temperatures)

#Max/Min Temperature: The highest and lowest temperatures in each city.
min_temperatures = np.min(temperature_data, 1)
print(min_temperatures)
max_temperatures = np.max(temperature_data, 1)
print(max_temperatures)

#Day Comparison: Temperature differences between Day 1 and Day 7 in each city.
day_comparison = [int(temperature_data[i, 0] - temperature_data[i, 6]) for i in range(temperature_data.shape[0])]
print(day_comparison)
```

1. Sampling and Shuffling Data: Perform random sampling on a dataset, followed by shuffling the sampled data.

Expected Outputs:

A randomly sampled subset and its shuffled version.
Calculated mean, median, variance, skewness, and kurtosis of the dataset.

```python
data = np.arange(100)
print(data)
sample_data = np.random.choice(data, size= 10, replace= False)
print(sample_data)
np.random.shuffle(sample_data)
print(sample_data)

average = np.mean(sample_data)
print(average)
median = np.median(sample_data)
print(median)
variance = np.var(sample_data)
print(variance)

from scipy.stats import skew, kurtosis
print("Skewness:", skew(sample_data))
print("Kurtosis:", kurtosis(sample_data))



```

1. Data Creation Exercise:

Objective: Create a simulated time series dataset.

Instructions:

Generate a date range from January 1, 2022, to December 31, 2022, with a daily frequency.
Simulate daily temperature data for this period. Assume temperature varies between 10°C and 35°C.
Simulate daily humidity data for the same period. Assume humidity varies between 30% and 90%.
Structure the data into a DataFrame with columns: Date, Temperature, and Humidity.
2. Data Analysis Exercise:

Objective: Perform basic data analysis on the time series data.

Instructions:

Calculate the mean, median, and standard deviation for both temperature and humidity.
Identify the date with the highest temperature and the date with the highest humidity.
[bonus] Plot the temperature and humidity trends over the year.
3. Window Function Exercise:

Objective: Apply window functions to the time series data.

Instructions:

Use a rolling window of 7 days to calculate the moving average for both temperature and humidity.
[bonus] Plot the original data and the 7-day moving average on the same graph for comparison.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating date ranges
dates = np.arange('2022-01-01', '2023-01-01', dtype='datetime64[D]')
temperatures = np.random.uniform(10, 35, 365)
humidity = np.random.uniform(30, 90, 365)

# Creating the DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperatures,
    'Humidity': humidity
})

print(df['Temperature'].mean())
print(df['Temperature'].median())
print(df['Temperature'].std())
print(df['Humidity'].mean())
print(df['Humidity'].median())
print(df['Humidity'].std())
print(df[df['Temperature']== df['Temperature'].max()]['Date'])
print(df[df['Humidity']== df['Humidity'].max()]['Date'])

# Set up the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Temperature as line plot
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature (°C)')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create second y-axis for Humidity
ax2 = ax1.twinx()

# Plot Humidity as histogram/bar plot
ax2.bar(df['Date'], df['Humidity'], width=3, alpha=0.3, color='blue', label='Humidity (%)')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title and layout
plt.title('Temperature and Humidity Trends Over the Year')
fig.tight_layout()
plt.show()

rolling_mean = df['Temperature'].rolling(window=7).mean()
print("Rolling Mean:\n", rolling_mean)
```
